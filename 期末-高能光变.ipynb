```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
data = pd.read_csv("ad60003000g200170m_g2_t1.csv")  

t =data['TIME']
counts = data['RATE']

plt.figure(figsize=(10,6))
plt.plot(t,counts,'k')
plt.xlabel("TIME(s)",fontsize=12)
plt.ylabel("RATE(coount/s)",fontsize=12)
plt.title("lightcurve",fontsize=15)
plt.show()  
```
